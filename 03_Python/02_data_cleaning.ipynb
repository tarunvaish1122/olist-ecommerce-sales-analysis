## Checking Missing Values

```python
datasets = {
    "customers":customers,
    "orders":orders,
    "order_items":order_items,
    "order_payments":order_payments,
    "order_reviews":order_reviews,
    "products":products,
    "product_category_name_translation":product_category_name_translation,
    "sellers":sellers,
    "geolocation":geolocation
    
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print(df.isnull().sum())
```

## Missing Values Analysis

After checking missing values across all datasets, the following observations were made:

### Orders Dataset

Missing values were found in:

- `order_approved_at`
- `order_delivered_carrier_date`
- `order_delivered_customer_date`

These missing values are expected because some orders are still in statuses such as **processing**, **invoiced**, or **shipped**. Since these orders were not delivered, delivery-related timestamps are naturally unavailable.

**Decision:** No imputation was performed. The missing values were retained.

---

### Order Reviews Dataset

Missing values were found in:

- `review_comment_title`
- `review_comment_message`

Many customers submitted only a review score without writing a text review.

**Decision:** The missing values were retained because they represent the absence of customer comments rather than missing data.

---

### Products Dataset

Missing values were found in:

- `product_category_name`
- `product_name_lenght`
- `product_description_lenght`
- `product_photos_qty`

These missing values are part of the original Olist dataset and represent incomplete product metadata.

Only **2 records** had missing values in `product_weight_g`.

**Decision:**

- Product metadata columns were left unchanged.
- Missing values in `product_weight_g` will be replaced using the median value because it is a numerical column with only two missing records.

```python
products["product_weight_g"] = products["product_weight_g"].fillna(
    products["product_weight_g"].median()
)
```

```python
products["product_category_name"] = (
    products["product_category_name"]
    .fillna("Unknown Category")
)
```

## Duplicate Records Check

The next step is to identify duplicate records in each dataset.

Duplicate records can affect aggregations, KPIs, and analytical results.

Each dataset is checked separately before deciding whether duplicate rows should be removed.

```python
for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()}")
```

## Duplicate Records Analysis

Duplicate records were checked across all datasets.

Observations:

- No duplicate records were found in Customers, Orders, Order Items, Order Payments, Order Reviews, Products, Sellers, and Product Category Translation.
- The Geolocation dataset contains duplicate records. This is expected in the original Olist dataset because multiple latitude and longitude observations are available for the same ZIP code.

**Decision:**

The duplicate records in the Geolocation dataset were retained because they represent valid geographic observations and are not used in the current business analysis.

## Data Types Analysis

Most columns have appropriate data types.

However, all date-related columns were imported as **object** data type because CSV files store dates as text.

For accurate date calculations and time-based analysis, these columns will be converted to the `datetime` data type.

```python
date_columns = {
    "orders": [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date"
    ],
    
    "order_items": [
        "shipping_limit_date"
    ],
    
    "order_reviews": [
        "review_creation_date",
        "review_answer_timestamp"
    ]
}
```

```python
for column in date_columns["orders"]:
    orders[column] = pd.to_datetime(orders[column])

for column in date_columns["order_items"]:
    order_items[column] = pd.to_datetime(order_items[column])

for column in date_columns["order_reviews"]:
    order_reviews[column] = pd.to_datetime(order_reviews[column])
```

```python
print(orders.dtypes)
print(order_items.dtypes)
print(order_reviews.dtypes)
```
